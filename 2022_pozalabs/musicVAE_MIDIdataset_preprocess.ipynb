```python
import music21 # MIDI 파일 조작을 위한 라이브러리
import numpy as np
import pandas as pd # 데이터 전처리 및 csv 파일 입출력
import os # 경로 설정
import pickle # 데이터 저장 라이브러리
import glob # 모든 하위 폴더의 파일에 접근
```

```python
data_path = "/content/drive/MyDrive/applications/2022pozalabs/data/groove/"
file_name = "drummer1/session1/100_neworleans-secondline_94_beat_4-4.mid"
```

```python
# 미디 파일 내부의 음악 정보 확인
midi_obj = music21.converter.parse(data_path + file_name)
midi_obj
```

```python
# 음악의 정보들이 저장된 오브젝트를 이용하여 노트의 음정과 시작지점 정보에 대한 array 출력
def create_mel_data_each_file(midi_obj):
    mel_data = dict()
    
#     print(len(midi_obj.flat.getElementsByClass(music21.chord.Chord)))
    for n in midi_obj.flat.getElementsByClass(music21.chord.Chord):
        if n.offset not in mel_data:
            mel_data[n.offset] = dict()
        mel_data[n.offset]['offset'] = n.offset
        for p in n.pitches:
            mel_data[n.offset]['note'] = p.midi

#     print(len(midi_obj.flat.getElementsByClass(music21.note.Note)))
    for n in midi_obj.flat.getElementsByClass(music21.note.Note):
        if n.offset not in mel_data:
            mel_data[n.offset] = dict()
        mel_data[n.offset]['offset'] = n.offset
        prev_p = 0
        for p in n.pitches:
            if prev_p < p.midi:
                mel_data[n.offset]['note'] = p.midi
            prev_p = p.midi    
    
    return mel_data
```

```python
# 한 곡에서 212개의 멜로디 노트 정보를 추출
mel_data = create_mel_data_each_file(midi_obj)
mel_data
```

```python
# 데이터셋의 모든 곡에 대한 노트 정보 추출
# 폴더 안의 모든 미디 파일명 리스트 생성
file_list = []

for file_name in os.listdir(glob(data_path+'*')):
    if file_name.endswith('.mid') or file_name.endswith('.midi'):
        file_list.append(data_path + file_name)
```
